```python
data_kotor = "   #001-wahyu aji-informatika-python,java,php-aktif   "

#PROSES

# 1. Menghapus spasi di awal dan akhir serta menghapus "#" di awal
data_bersih = data_kotor.strip().replace("#", "")
# strip() digunakan untuk menghapus spasi di awal dan diakhir
# replace() digunakan untuk menghapus tanda #

# 2. Memisahkan data atau memecah string berdasarkan tanda ("-")
data_split = data_bersih.split("-")
# split() digunakan untuk memecah string menjadi beberapa bagian atau list berdasarkan pemisah tertentu seperti tanda (-) pada data kotor tugas ini

# 3. Menyimpan ke variabel masing-masing
id_user = data_split[0]
nama = data_split[1].title()
prodi = data_split[2].upper()
skills = data_split[3].lower()
status = data_split[4].title()
# Pada bagian ini hasil split() disimpan ke dalam variabel sesuai kategori menggunakan posisi indeks masing-masing seperti contoh: nama[1]
# title() digunakan untuk membuat huruf awal jadi kapital
# upper() digunakan untuk membuat semua huruf jadi besar

# 4. Mengolah skill
list_skill = skills.split(",")
skill_format = " & ".join(list_skill)
# split(",") digunakan untuk memecah data skill yang dipisah tanda koma menjadi list
# join() digunakan untuk menggabungkan elemen dalam list menjadi satu string dengan pemisah tertentu seperti tanda (&) pada list ini

# 5. Menghitung Panjang ID dan jumlah skill
panjang_id = len(id_user)
jumlah_skill = len(list_skill)
# len() digunakan untuk menghitung jumlah karakter atau jumlah data

#OUTPUT

print(f"Data Mentah : '{data_kotor}'")

print("=" * 45)
print("        HASIL VERIFIKASI DATA")
print("=" * 45)

print(f"ID pendaftar   :{id_user} ({panjang_id} digit)")
print(f"Nama Lengkap   :{nama}")
print(f"Program Studi  :{prodi}")
print(f"Skill Dikuasai :{skill_format}")
print(f"Total Skill    :{jumlah_skill}")
print(f"Status Akun    :{status}")

print("=" * 45)
# print(f"...") atau formatted string (f-string) digunakan untuk menyisipkan variabel ke dalam string secara langsung tanpa perlu cara yang rumit
```
